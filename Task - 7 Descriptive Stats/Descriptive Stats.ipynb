Finding the Correlation with out Library In-Built functions

```python
def standard_deviation(x):
    length = len(x)
    mean = sum(x) / len(x)
    total = 0
    for i in x:
        a = (i - mean) ** 2
        total = total + a
    std = (total / length) ** 0.5
    return std
```

```python
def covariance(x,y):
    total = 0
    mean_x = x.mean()
    mean_y = y.mean()
    for i in range(len(x)):
        sum_x_y = (x[i] - mean_x) * ( y[i] - mean_y) 
        total = total + sum_x_y
    cov = total / len(x)
    return cov
```

```python
covariance(df['Mthly_HH_Income'],df['Annual_HH_Income'])
```

```python
def correlation(x,y):
    cov = covariance(x,y)
    x_std = standard_deviation(x)
    y_std = standard_deviation(y)
    corr = cov / (x_std * y_std)
    return corr
```

```python
correlation(df["Mthly_HH_Income"],df["Annual_HH_Income"])
```

# 8. Skewness

Definition : In probability theory and statistics, skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. The skewness value can be positive, zero, negative, or undefined. 

### Positive Skewness :
It is also called as the Right Skewness , as it says if the distribution curse has a long extended range in the right side of the mean .

#### Example for Positive Skewness
If we take the average income of the people in India. We can find that most of the people are belongs to Middle Class families so the frequency of less income is very high and the people with highest income are very less but there is many people more than the average range of people .

##### Mean , Median and Mode in Postive Skewness

Mode < Median < Mean

### Negative Skewness:
It is also called as Left Skewness , as it says if the distribution curse has a long extended range in the left side of the mean

## Example for  Negative Skewness:
If we consider the distribution of Age who died due to Normal Causes. 
There will be very few prople who dies in the child age or Adult age due to normal death Causes. But there is very high frequency at the age of 60 and 70 or more due to Health Issues and Organ Failures.

#### Mean, Median and Mode in Negative Skewness

Mode > Median > Mean

# 9.QQ Plot

Definition : In statistics, a QQ (quantile-quantile) plot is a probability plot, which is a graphical method for comparing two probability distributions by plotting their quantiles against each other.
1. It Checks the data Whether it is normally Distributed or not.
2. if the data falls in 45 degree Angle , then it is Noramally Distributed

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```
